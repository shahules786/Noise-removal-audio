## Using pydub

```python
from pydub import AudioSegment, silence
from scipy.io import wavfile
import numpy as np
```

```python
%config Completer.use_jedi = False
```

```python
def get_backgroundnoise(file,threashold=-32):
    
    audio = AudioSegment.from_wav(file)
    silences = silence.detect_silence(audio_segment=audio,min_silence_len=500, silence_thresh=threashold)
    longest_silences= np.argsort([start-end for start,end in silences])[0:2]
    longest_silences = [silences[x] for x in longest_silences ]
    concatenated = audio[longest_silences[0][0]:longest_silences[0][1]] + audio[longest_silences[1][0]:longest_silences[1][1]]
    
    concatenated.export("noise.wav",format='wav')
    
```
